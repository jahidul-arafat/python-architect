# Challenge_01: Factorial

```python
# function to calculate factorial of a number
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)
```

```python
print(factorial(5))
```

```python
factorial(1000)
```

```python
# function to calculate factorial of a number using iteration
def factorial_iter(n):
    # base cases
    if n == 0:
        return 1
    if n < 0 or type(n)!= int:
        return None
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
```

```python
factorial_iter(-3.2)
```

```python
# function to calculate factorial of a number using recursion and print each steps
def factorial_rec(n):
    # base cases
    if n == 0:
        return 1
    if n < 0 or type(n)!= int:
        return None
    print(n, end="*" if n > 1 else "")
    return n * factorial_rec(n - 1)
```

```python
factorial_rec(3)
```
